* what is DataFrame 
1. A DataFrame is a two-dimensional, tabular data structure in the pandas library (Python) that consists of rows and columns, similar to a spreadsheet or SQL table.
2. It allows for easy manipulation, analysis, and storage of structured data with labeled axes.

```python
# using a list
data=pd.DataFrame([20,39,40],columns=['age'])
data
```
